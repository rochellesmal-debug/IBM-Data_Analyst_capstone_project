**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')
df.to_csv('your_filename.csv', index=False)
# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Demo: Running an SQL Query


Count the number of rows in the table named 'main'


```python
QUERY = """
SELECT COUNT(*) 
FROM main
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
QUERY = "SELECT CompTotal FROM main WHERE CompTotal IS NOT NULL AND CompTotal < 1000000"
df_comp = pd.read_sql_query(QUERY, conn)
plt.hist(df_comp['CompTotal'],edgecolor='black')
plt.title('Distribution of Total Compensation')
plt.xlabel('Compensation')
plt.ylabel('Frequency')
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
age_map = {'Under 18 years old': 16, '18-24 years old': 21, '25-34 years old': 30, 
           '35-44 years old': 40, '45-54 years old': 50, '55-64 years old': 60, '65 years or older': 70}
df['AgeNumeric'] = df['Age'].map(age_map)

df.boxplot(column='AgeNumeric', figsize=(8, 6))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
plt.scatter(df['AgeNumeric'], df['WorkExp'], alpha=0.5)
plt.title('Scatter Plot: Age vs Work Experience')
plt.xlabel('Age')
plt.ylabel('Work Experience (Years)')
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
time_map = {'Less than 15 minutes a day': 7.5, '15-30 minutes a day': 22.5, '30-60 minutes a day': 45.0, 
            '60-120 minutes a day': 90.0, 'Over 120 minutes a day': 150.0}
df['TimeSearching_num'] = df['TimeSearching'].map(time_map)
df['Frustration_num'] = df['Frustration'].fillna('').apply(lambda x: len(x.split(';')) if x != 'None of these' else 0)
plt.scatter(df['TimeSearching_num'], df['Frustration_num'], s=df['AgeNumeric']*2, alpha=0.5)
plt.title('Bubble Plot: Time Searching vs Frustration')
plt.xlabel('Time Searching')
plt.ylabel('Frustration Count')
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
db_series = df['DatabaseWantToWorkWith'].dropna().str.split(';').explode()
db_counts = db_series.value_counts().head(5)
plt.pie(db_counts, labels=db_counts.index, autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents')
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
time_map = {
    'Less than 15 minutes a day': 7.5,
    '15-30 minutes a day': 22.5,
    '30-60 minutes a day': 45.0,
    '60-120 minutes a day': 90.0,
    'Over 120 minutes a day': 150.0
}
df['TimeSearching_num'] = df['TimeSearching'].map(time_map)
df['TimeAnswering_num'] = df['TimeAnswering'].map(time_map)
df_30_35 = df[df['Age'].isin(['25-34 years old', '35-44 years old'])]
medians = df_30_35[['TimeSearching_num', 'TimeAnswering_num']].median()
plt.figure(figsize=(3, 6))
plt.bar(['Median Time'], [medians['TimeSearching_num']], label='Time Searching')
plt.bar(['Median Time'], [medians['TimeAnswering_num']], 
        bottom=[medians['TimeSearching_num']], label='Time Answering')
plt.title('Time Searching and Answering (Age Group 30-35)')
plt.ylabel('Mintes')
plt.legend()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
median_comp = df[df['CompTotal'] < 1000000].groupby('AgeNumeric')['CompTotal'].median().reset_index()
subset = median_comp[(median_comp['AgeNumeric'] >= 45) & (median_comp['AgeNumeric'] <= 60)]

plt.plot(subset['AgeNumeric'], subset['CompTotal'], marker='o')
plt.title('Median Total Compensation for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median CompTotal')
plt.grid(True)
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
df['MainBranch'].value_counts().plot(kind='barh')
plt.title('Main Branch Distribution')
plt.xlabel('Count')
plt.show()
```
